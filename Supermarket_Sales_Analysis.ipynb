**Total Tax by Branch**

```python
branch_tax = df.groupby('Branch')['Tax 5%'].sum().reset_index()
branch_tax = branch_tax.sort_values(by='Tax 5%', ascending=False)
branch_tax
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Branch', y='Tax 5%', data=branch_tax)
plt.title('Total Tax per Branch')
plt.xlabel('Branch')
plt.ylabel('Total Tax')
plt.show()
```

```python
products_sales = df.groupby('Product line')['gross income'].sum().reset_index()
products_sales = products_sales.sort_values(by='gross income', ascending=False)
products_sales
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Product line', y='gross income', data=products_sales)
plt.title('Total Profit by Product Line')
plt.xlabel('Product Line')
plt.ylabel('gross income')
plt.xticks(rotation=45)
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Total'].sum().reset_index()
gender_sales = gender_sales.sort_values(by='Total', ascending=False)
gender_sales
```

```python
plt.figure(figsize=(6,6))
plt.pie(gender_sales['Total'], labels=gender_sales['Gender'], autopct='%1.1f%%', startangle=140)
plt.title('Total Sales by Gender')
plt.axis('equal')
plt.show()
```

```python
gender_avg = df.groupby('Gender')['Total'].mean().reset_index()
gender_avg
```

```python
gender_count = df['Gender'].value_counts().reset_index()
gender_count.columns = ['Gender', 'Count']
gender_count
```

```python
product_gender = df.groupby(['Gender','Product line'])['Total'].sum().reset_index()
product_gender = product_gender.sort_values(by='Total', ascending=False)
product_gender
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=product_gender)
plt.title('Product Sales by Gender')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
